```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import svm
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/content/Cancer.csv')
df
```

```python
df.drop(columns=['Unnamed: 32','id'],axis=1,inplace = True)
x = df.drop('diagnosis',axis=1)
df = df.replace({'M':1, 'B':0})
y = df.diagnosis
df.head(5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
df.corr()['diagnosis'].sort_values()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(max_depth = 4,random_state = 101)
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)    
```

```python
n_estimator = [64,100,150,200]
max_features = [2,3,4]
bootstrap = [True, False]
obb_score = [True, False]
```

```python
pram_grid = {'n_estimator':n_estimator,
             'max_features': max_features,
             'bootstrap':bootstrap,
             'obb_score':obb_score}
```

```python
rfc = RandomForestClassifier()
rfc = GridSearchCV(rfc,pram_grid)
```
